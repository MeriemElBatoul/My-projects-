```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from keras.optimizers import Adam
import warnings
warnings.filterwarnings("ignore")
```

```python
x = data.drop(columns=['Absence'])  # Caractéristiques (toutes les colonnes sauf la cible)
y = data['Absence']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
model=Sequential()
```

```python
model.add(Dense(11,activation= "relu",input_dim=11))
model.add(Dense(11,activation= "relu"))
model.add(Dense(11,activation= "relu"))
model.add(Dense(11,activation= "relu"))
model.add(Dense(11,activation= "relu"))
model.add(Dense(11,activation= "relu"))
model.add(Dense(1,activation= "sigmoid"))

model.compile(optimizer= "adam",loss="binary_crossentropy", metrics=['accuracy'])
model.summary()
```

```python
history= model.fit(x_train,y_train,epochs=30,validation_split=0.3)
```

```python
tr_loss = history.history['loss']
val_loss = history.history['val_loss']
index_loss = np.argmin(val_loss)
val_lowest = val_loss[index_loss]

Epochs = [i+1 for i in range(len(tr_loss))]
loss_label = f'best epoch= {str(index_loss + 1)}'
plt.figure(figsize= (20, 8))
plt.style.use('fivethirtyeight')

plt.plot(Epochs, tr_loss, 'r', label= 'Training loss')
plt.plot(Epochs, val_loss, 'g', label= 'Validation loss')
plt.scatter(index_loss + 1, val_lowest, s= 150, c= 'blue', label= loss_label)
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout
plt.show()
```

```python
y_pred=model.predict(x_test)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import r2_score,accuracy_score

R2 = r2_score(y_test, y_pred)
print("R2 Score=",R2 )
```

```python
from sklearn.metrics import confusion_matrix
Accurcy = accuracy_score(y_test, y_pred)
print("Accurcy=",Accurcy )
```

```python
from sklearn.metrics import classification_report, confusion_matrix
conf_matrix=confusion_matrix(y_test,y_pred)
conf_matrix
sns.heatmap(conf_matrix,annot=True, fmt='d')
plt.show()
```

```python
print(f"Classification report:\n{classification_report(y_test, y_pred)}")
print("")
print("_"*12)
print("")
```

The model is highly skewed towards classifying the majority class (0). It achieves a 100% recall for class 0 but fails completely for class 1 (which has a precision and recall of 0.00).
This is a clear indication of class imbalance, as the ANN is overfitting on the majority class and ignoring the minority class (1).
we need to apply **class-balancing** techniques like  class weighting to improve its performance on minority cases
